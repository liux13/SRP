```python
from contextlib2 import closing
```

```python
def get_sample(train):
    return gtp.random_sample(train = train)
```

```python
def get_batch(bsize, train):
    data = np.empty((bsize, 6, 64, 64))
    xy = np.empty((bsize, 2))
    cs = np.empty((bsize,2))
    label = np.empty((bsize, 1))
    
    for i in range(bsize):
        sample = get_sample(train)
        data[i] = sample.data
        xy[i] = sample.predicted.dx, sample.predicted.dy
        R = np.radians(sample.predicted.angle)
        cs[i] = np.cos(R), np.sin(R)
        label[i] = sample.predicted.label
    return (data, xy, cs, label)
```

```python
def unflatten(a):
    data = a[:128*6*64*64].reshape(128, 6, 64, 64)
    a = a[data.size:]
    xy = a[:128*2].reshape(128,2)
    a = a[128*2:]
    cs = a[:128*2].reshape(128,2)
    a = a[128*2:]
    label = a[:128]
    return (data, xy, cs, label)

SAMPLE_SIZE = 128*(6*64*64 + 2 + 2 + 1)


def flatten(data, xy, cs, label):
    a = np.empty(SAMPLE_SIZE)
    b = a
    b[:128*6*64*64] = data.flat
    b = b[128*6*64*64:]
    b[:128*2] = xy.flat
    b = b[128*2:]
    b[:128*2] = cs.flat
    b = b[128*2:]
    b[:128] = label.flat
    return a
    
    
```
